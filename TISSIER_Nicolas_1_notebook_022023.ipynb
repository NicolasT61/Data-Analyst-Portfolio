# la proportion de personnes en état de sous-nutrition

```python
# Garder l'année 2017 et mettre en millions les valeurs du CSV population
df_pop_2017=population
#df_pop_2017['Année'].unique()

df_pop_2017.drop(df_pop_2017[(df_pop_2017['Année'] != 2017)].index, inplace=True)
df_pop_2017['Valeur'] = df_pop_2017['Valeur'].multiply(0.001)
df_pop_2017
```

```python
#sous_nutrition.info()
print(sous_nutrition['Zone'].nunique())
print(sous_nutrition['Année'].unique())
print(sous_nutrition['Valeur'].unique())
```

```python
# Garder l'année 2017, mettre toutes les valeurs en float et supprimer les NaN
df_sous_nutrition=sous_nutrition
df_sous_nutrition.drop(df_sous_nutrition[(df_sous_nutrition['Année'] != '2016-2018')].index, inplace=True)
df_sous_nutrition.replace('2016-2018','2017',inplace=True)
df_sous_nutrition.replace('<0.1','0.05',inplace=True)
df_sous_nutrition.dropna(inplace=True)

df_sous_nutrition["Valeur"] = pd.to_numeric(df_sous_nutrition["Valeur"])
df_sous_nutrition.sort_values(by = 'Valeur', ascending = False).head(5)
```

```python
#Fusion de la population totale 2017 avec la population en sous nutrition en fonction des pays, suppression des colonnes Année et changement de noms des valeurs
df_pop_sous_nutri=df_pop_2017.merge(df_sous_nutrition,how='inner', on = ['Zone'])
df_pop_sous_nutri.drop(['Année_x','Année_y'],axis=1,inplace=True)
df_pop_sous_nutri.rename(columns = {'Valeur_x': 'Population totale', 'Valeur_y': 'Population en sous nutrition'},inplace=True)
df_pop_sous_nutri["% de la population en sous nutrition"]=(df_pop_sous_nutri['Population en sous nutrition'] / df_pop_sous_nutri['Population totale'] *100)
df_pop_sous_nutri.loc[df_pop_sous_nutri['Population en sous nutrition'] > 0.05].sort_values(by = '% de la population en sous nutrition', ascending = False).head(20)
#df_pop_sous_nutri.sort_values(by = '% de la population en sous nutrition', ascending = False)
# Ne pas afficher les 0.05 faire un unique pour voir si existzant
```

```python
df_pop_sous_nutri.dtypes
```

```python
Population_total = df_pop_sous_nutri['Population totale'].sum()
Population_sous_nutrition=df_pop_sous_nutri['Population en sous nutrition'].sum()
proportion=(Population_sous_nutrition / Population_total)*100

print ("En 2017 les pays où la sous-nutrition existe, le nombre d'habitant total est de ",  round(Population_total,2),  "millions. Le nombre de personne en sous nutrition est de " , Population_sous_nutrition , "millions, soit " , proportion , "%")
```

# l’utilisation de la disponibilité intérieure, en particulier la part qui est attribuée à l’alimentation animale, celle qui est perdue et celle qui est concrètement utilisée pour l'alimentation humaine. Je crois que Julien avait trouvé un moyen de facilement calculer ces proportions.

```python
dispo_alimentaire_new
disponibilite_interieur_vs_animaux= dispo_alimentaire_new.drop(columns=['Disponibilité alimentaire (Kcal/personne/jour)',	'Disponibilité alimentaire en quantité (kg/personne/an)',	'Disponibilité de matière grasse en quantité (g/personne/jour)',	'Disponibilité de protéines en quantité (g/personne/jour)'])
disponibilite_interieur_vs_animaux.fillna(0, inplace = True)
disponibilite_interieur_vs_animaux['Dispo_alim'] = disponibilite_interieur_vs_animaux['Production'] + disponibilite_interieur_vs_animaux['Importations - Quantité'] - disponibilite_interieur_vs_animaux['Exportations - Quantité'] + disponibilite_interieur_vs_animaux['Variation de stock']
disponibilite_interieur_vs_animaux['Dispo_alim_humain'] = disponibilite_interieur_vs_animaux['Dispo_alim'] - disponibilite_interieur_vs_animaux['Aliments pour animaux'] - disponibilite_interieur_vs_animaux['Semences'] - disponibilite_interieur_vs_animaux['Traitement'] - disponibilite_interieur_vs_animaux['Autres Utilisations'] - disponibilite_interieur_vs_animaux['Pertes']
#disponibilite_interieur_vs_animaux

ratio_disp_interieur_vs_animal=(disponibilite_interieur_vs_animaux['Aliments pour animaux'].sum()/disponibilite_interieur_vs_animaux['Disponibilité intérieure'].sum())*100
print('La part qui est attribuée à l’alimentation animale représente' , ratio_disp_interieur_vs_animal,' %' ,"de l'utilisation de la disponibilité intérieur")

ratio_disp_interieur_vs_pertes=(disponibilite_interieur_vs_animaux['Pertes'].sum()/disponibilite_interieur_vs_animaux['Disponibilité intérieure'].sum())*100
print('La part qui est attribuée aux pertes représente' , ratio_disp_interieur_vs_pertes,' %' ,"de l'utilisation de la disponibilité intérieur")

ratio_disp_interieur_vs_dispo_alim_humain_1=(disponibilite_interieur_vs_animaux['Dispo_alim_humain'].sum()/disponibilite_interieur_vs_animaux['Disponibilité intérieure'].sum())*100
print('La part qui est attribuée à l’alimentation humaine représente' , ratio_disp_interieur_vs_dispo_alim_humain_1,' %' ,"de l'utilisation de la disponibilité intérieur")


```
